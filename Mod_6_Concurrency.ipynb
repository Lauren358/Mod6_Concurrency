```python
'''13.1 Write the current date as a string to the text file today.txt.'''

from datetime import date
now = date.today()
now_string = now.isoformat()
with open('today.txt', 'wt') as output:
    print(now_string, file=output)
```

```python
'''13.2 Read the text file today.txt into the string today_string.'''
with open('today.txt', 'rt') as input:
    today_string = input.read()

today_string
```

```python
'''13.3 Parse the date from today_string.'''
from datetime import datetime

fmt = '%Y-%m-%d\n'
datetime.strptime(today_string, fmt)
```

```python
'''15.1 Use multiprocessing to create three separate processes. 
Make each one wait a random number of seconds between zero and one, print the current time, and then exit.'''

import multiprocessing
import time
import os

def now(seconds):
    from datetime import datetime
    from time import sleep
    sleep(seconds)
    print('wait', seconds, 'seconds, time is', datetime.datetime())

if __name__ == '__main__':
    import random
    for n in range(3):
        seconds = random.random()
        exc = multiprocessing.Process(target=now, args=(seconds,))
        exc.start()
```
